# Task 1

```python
d = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
```

```python
text = 'oct-08_singapore.txt'

month_date, city_ext = text.split('_')
month,date = month_date.split('-')
month = d[month]

new_text = f'{month}-{date}-{city_ext}'
new_text
```

# Task 2

```python
def rename_my_file(old_file):
    month_date, city_ext = old_file.split('_')
    month,date = month_date.split('-')
    month = d[month]

    new_name = f'{month}-{date}-{city_ext}'
    return new_name
```

```python
for i in d1:
    new_name = rename_my_file(i)
    print(f"{i} ---> {new_name}")
```

```python
import os
import shutil 

directory = '/Users/annguyen/Documents/GitHub/learning-portfolio-myboomie/files, folders & os  exercises/os-collaboration-exercise-data'
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities:
    city_path = os.path.join(directory, city)
    new_path = os.path.join(directory,city,'test.txt')
    if not os.path.exists(new_path):
        shutil.copy('test.txt', new_path)
    print(f"Copied to {new_path}")
    
```
