# 2 Comparing the two ‘dialects’

### pyplot Interface

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create the plot using pyplot
plt.plot(x, y)
plt.title("Simple Plot Using Pyplot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### OO Interface

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot data on the axes
ax.plot(x, y)
ax.set_title("Simple Plot Using Object-oriented Interface")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Display the plot
plt.show()

```

**Observation**:
- Both `pyplot` and OO interface can produce the same figure
- Both have some similar syntax such as:
|Purpose|`pyplot`|`OO interface`|
|:--:|:--:|:--:|
|Show a plot|`plt.show()`|`plt.show()`|
- However, they have same same but different syntax; for example:
|Purpose|`pyplot`|`OO interface`|
|:--:|:--:|:--:|
|Set title|`plt.title()`|`ax.set_title()`|
|Set y-axis label|`plt.ylabel()`|`ax.set_ylabel()`|
|Set x-axis label|`plt.xlabel()`|`ax.set_xlabel()`|
|Start plotting|`plt.plot()`|`ax.plot()`|
|Start a figure|`plt.plot()`|`fig, ax = plt.subplots()`|


# 3 What is OO ax, really?

In using OO interface, we have to start with `fig, ax = plt.subplots()`

Here is the meaning and syntax of `plt.subplots()`:

`matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, width_ratios=None, height_ratios=None, ...)`

By default, `nrows` and `ncols` if not specified would be 1. The `nrows` refer to the number of rows that our figure would have, similar for `ncols`. If a single plot is desired, meaning our figure would be in the row 1 and col 1. If two side-by-side plots are required, it means we will need a total of 1 rows (`nrows = 1`) and 2 columns (`ncols = 2`). One plot is in (row 1, col 1); and the other will be in (row 1, col 2).
![](https://python-course.eu/images/numerical-programming/creating-subplots-in-matplotlib_6.webp)

**In other words, `plt.subplots()` will create a 2D array for us in which each position is dedicated for one figure/plot. To access each figure/plot, we can call them by its position via `[row, col]` (similar to indexing in array)**

For example, with `fig, ax = plt.subplots(2,2)` like the above visual demonstration, we can go into the figure with axis [1,1] to perform our visualisation:

```python
#ax[1,1].set_xlabel(....)
```

# 4 A complete OO example

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 2 * np.pi, 500)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5  # Second dataset for comparison

# Create a figure and a set of subplots
fig, (ax1, ax2) = plt.subplots(2, 1)  # 2 rows, 1 column

# First subplot: Fill between y1 and y2
ax1.plot(x, y1, 'r', alpha=0.7)  # First line
ax1.plot(x, y2, 'g', alpha=0.7)  # Second line
ax1.fill_between(x, y1, y2, color='gray', alpha=0.3)  # Fill between the two lines
ax1.set_title('Fill Between Lines')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# Second subplot: Fill under the curve
ax2.plot(x, y1, 'b', alpha=0.7)  # Plot the line
ax2.fill_between(x, 0, y1, color='blue', alpha=0.1)  # Fill under the curve
ax2.set_title('Fill Under Curve')
ax2.set_xlabel('x')
ax2.set_ylabel('sin(x)')

# Set the overall figure title
fig.suptitle('Comparison of Two Datasets with Filled Areas', fontsize=16)

# Automatically adjust subplot parameters to give specified padding
fig.tight_layout(pad=3.0)

plt.show()
```

### Some functions syntax

`ax.fill_between()`: `Axes.fill_between(x, y1, y2=0, where=None, interpolate=False, step=None, *, data=None, **kwargs)` where:
- we will fill the area between two horizontal curves.

- The curves are defined by the points `(x, y1)` and `(x, y2)`. This creates one or multiple polygons describing the filled area.

- We may exclude some horizontal sections from filling using `where`.

For example, in the code: `ax1.fill_between(x, y1, y2, color='gray', alpha=0.3)`

Here, we are to fill the area under area of $y_1=sin(x)$ and $y_2 = sin(x) + 0.5$ in the range of $x$ from $[0, 2\pi]$ with color `gray` with transparency degree of `0.3`

Or, in the code: `ax2.fill_between(x, 0, y1, color='blue', alpha=0.1)`

Here, we are to fill the area under the area of $y=0$ and $y=y_1=sin(x)$ in the range of $x$ from $[0, 2\pi]$ with the color `blue` and transparency degree of `0.1`

`fig.tight_layout()`:`matplotlib.pyplot.tight_layout(*, pad=1.08, h_pad=None, w_pad=None, rect=None)` where:
- padfloat, default: 1.08
- Padding between the figure edge and the edges of subplots, as a fraction of the font size

# 5 Multiple rows and columns

### Using ax

```python
# Your code here
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

### Some functions syntax

`ax.flatten()`: recall that `ax` is a 2D array of size (`nrow`, `ncol`). When we flatten array, we will do it row by row, column by column.

For example, if ax is (2,2), when flattening it we would have: `[0,0]`, `[0,1]`, `[1,0]`, `[1,1]` which corresponds to its position of top_left, top_right, bottom_left, bottom_right respectively.



`bottom_right.hlines(y=4, xmin=0, xmax=5)` :`Axes.hlines(y, xmin, xmax, colors=None, linestyles='solid', label='', *, data=None, **kwargs)`

=> Plot horizontal lines at each y from xmin to xmax.

- Parameters:
    - **y**: float or array-like; indexes where to plot the lines.

    - **xmin, xmax**: float or array-like, Respective beginning and end of each line. If scalars are provided, all lines will have the same length.

## 5.1 Sharing axes

`plt.subplots()`: `matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, width_ratios=None, height_ratios=None, subplot_kw=None, gridspec_kw=None, **fig_kw)[source]`
**Parameters**:
- **nrows**, **ncolsint**, default: 1
    - Number of rows/columns of the subplot grid.

- **sharex**, **shareybool** or {'none', 'all', 'row', 'col'}, default: False
    - Controls sharing of properties among x (sharex) or y (sharey) axes:

    - `True` or `'all'`: x- or y-axis will be shared among all subplots.

    - `False` or `'none'`: each subplot x- or y-axis will be independent.

    - `'row'`: each subplot row will share an x- or y-axis.

    - `'col'`: each subplot column will share an x- or y-axis.

    - When subplots have a shared x-axis along a column, only the x tick labels of the bottom subplot are created. Similarly, when subplots have a shared y-axis along a row, only the y tick labels of the first column subplot are created.

```python
# if sharex = True, sharey = 'row'
# x-axis will be shared among all subplots, here, we can see that all plots similar x-axis range
# since sharey = 'row', all the subplots in the same row will share the same y axis. 
# Here, subplots [0,1] and [0,0] share the same y-axis while [1,0] and [1,1] share the same y-axis

# Your code here
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex=True, sharey='row')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

### A quick helper function

```python

# Your code here
def get_plot_data():
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1)

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    return data
```

### The histogram

```python

# Your code here
plt.style.use('bmh')
data = get_plot_data()

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS'])
plt.xlabel('Mean of Basic Monthly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
```

## 6.2 Scatter plots: `plt.scatter(x,y,...)`

```python
plt.style.use("seaborn-v0_8-darkgrid")

data = get_plot_data()

for label, numbers in data.items():
    x = range(len(numbers))
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend()
```

## 6.3 Bar charts: `plt.bar(x,y,..)`

```python
# Data: Average daily temperatures for a week
average_daily_temperatures = {
    "Monday": 22,
    "Tuesday": 24,
    "Wednesday": 19,
    "Thursday": 21,
    "Friday": 23,
    "Saturday": 25,
    "Sunday": 20
}

# Extracting keys and values from the dictionary
days = list(average_daily_temperatures.keys())
temperatures = list(average_daily_temperatures.values())

# Creating the bar graph
plt.bar(days, temperatures, color='skyblue')

# Adding the title and labels with units
plt.title('Average Daily Temperatures for a Week')
plt.xlabel('Day of the Week')
plt.ylabel('Temperature (°C)')

# Adding data labels on top of each bar
for i, temp in enumerate(temperatures):
    plt.text(i, temp + 0.5, f'{temp}°C', ha='center')

# Display the plot
plt.xticks(rotation=45)  # Rotates x-axis labels to prevent overlap
plt.tight_layout()  # Adjusts subplot params so that the subplot(s) fits in to the figure area
plt.show()
```

**Add text/annotation on figure: `plt.text()`**

Syntax:
    `matplotlib.pyplot.text(x, y, s, fontdict=None, **kwargs)`
Purpose:
- Add text to the Axes.

- Add the text `s` to the Axes at location `x`, `y` in data coordinates, with a **default horizontal** alignment on the left and vertical alignment at the baseline.

Parameters:
- `x`, `y`: float
    - The position to place the text. By default, this is in data coordinates. The coordinate system can be changed using the transform parameter.
- `s`: str
    - The text

## 6.4 Pie charts: `plt.pie()`

```python
import matplotlib.pyplot as plt

# Data: Market share of different brands
market_share = {
    "Brand A": 25,
    "Brand B": 35,
    "Brand C": 20,
    "Brand D": 15,
    "Brand E": 5
}

# Labels and sizes for the pie chart
labels = market_share.keys()
sizes = market_share.values()

# Creating the pie chart
plt.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=0)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Adding a title
plt.title('Market Share of Different Brands')

# Display the pie chart
plt.show()
```
