## 1.2 Adding another plot

We can add multiple plots in the current figure by simplying adding another `plt.plot(x,y,...)`

```python
import numpy as np
x = np.arange(1,10,1)
y1 = x*2+3
y2 = x**2+3

plt.plot(x,y1, label ='$ y = 2x + 3$')
plt.plot(x,y2, label ='$ y = x^2 + 3$')
plt.legend()
plt.title('Some mathematical plots')
plt.grid(alpha = .25)
plt.show()
```

## 1.3 Yet another plot but with error bars: `plt.errorbar(x, y, xerr, yerr, ...)`

```python
import numpy as np
x = np.arange(1,10,1)
y1 = x*2+3
y2 = x**2+3
y3 = x**3 + 3
x_err, y_err = .25, .75

plt.plot(x,y1, label ='$ y = 2x + 3$')
plt.plot(x,y2, label ='$ y = x^2 + 3$')
plt.errorbar(x,y3, xerr = x_err, yerr = y_err, c = 'green', label = 'With error bars')
plt.legend()
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Some mathematical plots')
plt.grid(alpha = .25)
plt.show()
```

# 2 Better with NumPy

## 2.1 Adding mathematical functions

```python
x = np.arange(-5,5,.01)
plt.plot(x, np.sin(x), label ='$ sin(x) $', color = 'red')
plt.plot(x,np.cos(x), label = '$ cos(x)$', color = 'blue')
plt.legend()
plt.grid(alpha = 0.5)
plt.title('Sin and cos functions')
plt.show()
```

## We need another axis!

Use `twinx()` to share x-axis but have another y-axis on the right. This can be useful when we have 2 line plots with different y-axis range but similar x-axis range

```python
x = np.arange(-5,5,.01)
plt.plot(x, np.sin(x), label ='$ sin(x) $', color = 'red')
plt.legend(loc = 'upper left')

plt.twinx()

plt.plot(x,np.cos(x), label = '$ cos(x)$', color = 'blue')
plt.legend()

plt.grid(alpha = 0.5)
plt.title('Sin and cos functions')
plt.show()
```

### Things to note

1. We now have two `legend()` calls, one for each axis.
2. Use `np.linspace()` to specify the number of data points within a specified range (included) we want

```python
xtry = np.linspace(0,10,5) 
#create an array that starts from 0 and ends with 10 in which there are 5 elements inside the array
xtry
```

# 3 Saving to disc: `plt.savefig()`

We can save the plot to our current folder by `plt.savefig()`, inside the parenthesis, we can choose to save the image with name `ABC` by specifying it as arguments in `plt.savefig('ABC')`

We can also specify resolution by use dots-per-inch (`dpi`): `plt.savefig('ABC', dpi = 150)`


If you want it saved elsewhere, specify the path in more detail

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = 'FEV.csv'
import pandas as pd
df = pd.read_csv(data)
df
```

```python
#plot the FEV 
plt.scatter(df['FEV'], df['Hgt'], color = 'red', label = 'Hgt vs FEV')
plt.legend()
plt.grid(alpha = .5)
plt.show()
```

Letâ€™s add a line to highlight the trend shown by the data. This can be done using Linear Regression in `seaborn` package

```python
#plot the FEV 
import seaborn as sb
sb.regplot(data = df,x = 'FEV', y = 'Hgt') #Add a straight line for trend

plt.show()
```

## 4.2 Styles

### xkcd!

```python
with plt.xkcd():
    plt.scatter(df['FEV'], df['Hgt'], color = 'red', label = 'Hgt vs FEV')
    plt.legend()
    plt.show()
```

### Resetting styles

```python
plt.style.use('default')
plt.scatter(df['FEV'], df['Hgt'], color = 'red', label = 'Hgt vs FEV')
plt.legend()
plt.show()
```
